```r
names(lm01)
```

```r
residuals(lm01)[1:10]
```

```r
RSS = sum(residuals(lm01)^2)
```

```r
RSS
```

```r
MSE = sqrt(RSS/198)
```

```r
MSE
```

```r
mean_sales = mean(Advertising$sales)
```

```r
mean_sales
```

```r
TSS = sum((Advertising$sales - mean_sales)^2)
```

```r
TSS
```

```r
R2 = 1 - RSS/TSS
```

```r
R2
```

```r
cor(Advertising$TV, Advertising$sales)^2
```

```r
cor(Advertising$sales, predict(lm01))^2
```

```r
p = 1
```

```r
n = 200
```

```r
R_adj = 1 - ((1 - R2) * (n - 1) / (n - p - 1))
```

```r
R_adj
```

```r
F_stat = ((TSS-RSS)/p)/(RSS/(n-p-1))
```

```r
F_stat
```

```r
p_value = pf(F_stat, p, n-p-1, lower.tail=FALSE)
```

```r
p_value
```

```r
str_interp("RSS = ${RSS}, MSE = ${MSE}, R2 = ${R2}, R_adj = ${R_adj}, F = ${F_stat}, 
           p = ${p_value}")
```
